### Univariate Analysis

```python
#Making function to visualize top 20 products and top 5 countries
def plot(df, column_name, n=20): #default n = 20, we can customize it.
    top = df[column_name].value_counts().head(n)
    ax = top.plot(kind='barh', figsize=(8, 6))
    plt.grid(False)

    # Adding value on each bar
    for i, v in enumerate(top):
        ax.text(v, i, str(v), ha='left', va='center')
```

```python
#OK -> dimasukkan ke laporan (output codenya saja)
#Showing top 20 of products
plot(df, 'Description', n=20)
plt.title("Top 20 Product with The Highest Buying")
```

Kode yang diberikan adalah sebuah fungsi Python yang bertujuan untuk memvisualisasikan 20 produk teratas berdasarkan kolom 'Description' dari DataFrame df. Fungsi ini menggunakan diagram batang horizontal untuk menampilkan informasi teratas, dengan setiap bar yang menunjukkan jumlah kemunculan setiap produk dalam data. output nya berupa produk mana yang paling sering muncul dalam data, sehingga memberikan gambaran visual tentang popularitas atau frekuensi pembelian setiap produk. ### Bivariate Analysis

```python
#OK -> dimasukkan ke laporan (output codenya saja)
# Visualization
plt.figure(figsize=(8, 6))
ax = sns.barplot(data=top10_product, x='Description', y='Transaction', palette="blend:#7AB,#EDA")
plt.xticks(rotation=90)
plt.xlabel("Product")
plt.title("Top 10 products with the Most Transactions")

# Adding value on each bar
for i, v in enumerate(top10_product['Transaction']):
    ax.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd

data1 = {
    'Country': ['USA', 'USA', 'USA', 'Canada', 'Canada', 'UK', 'UK', 'Canada'],
    'total_price': [100, 150, 200, 50, 75, 120, 180, 90],
    'total_customer': [5, 3, 6, 2, 4, 4, 5, 3]
}

df1 = pd.DataFrame(data1)
```

```python
country_user1 = (df1
              .groupby(['Country'])
              .agg(total_spend=('total_price', 'sum'),
                   total_customer=('total_customer', 'nunique'))
              .reset_index())
print(country_user1)
```

Kode di atas untuk membuat DataFrame menggunakan pustaka pandas di Python. DataFrame ini berisi informasi tentang total harga (total_price) dan jumlah pelanggan (total_customer) untuk setiap negara tertentu. Setelah DataFrame dibuat, dilakukan pengelompokan berdasarkan negara, diikuti dengan operasi agregasi untuk menghitung total pengeluaran dan jumlah pelanggan unik di setiap negara.

Output dari kode ini adalah DataFrame country_user1 yang menunjukkan total pengeluaran (total_spend) dan jumlah pelanggan unik (total_customer) di setiap negara. DataFrame ini akan berisi tiga kolom: 'Country', 'total_spend', dan 'total_customer', yang masing-masing menunjukkan negara, total pengeluaran, dan jumlah pelanggan unik di negara tersebut. Hal ini dapat membantu dalam memahami pola pengeluaran dan demografi pelanggan berdasarkan negara, yang dapat berguna dalam analisis RFM.

```python
#OK -> dimasukkan ke laporan (output codenya saja)
customer_data['RecencyScore'] = pd.cut(customer_data["Recency"],
                                       bins=[-1,
                                             np.percentile(customer_data["Recency"], 25),
                                             np.percentile(customer_data["Recency"], 50),
                                             np.percentile(customer_data["Recency"], 75),
                                             #np.percentile(customer_data["Recency"], 95),
                                             customer_data["Recency"].max()],
                                       labels=[4, 3, 2, 1]).astype("int")
customer_data["RecencyScore"].value_counts()
# 4 artinya customernya itu memiliki skor tertinggi karena aktivitas pembeliannya memiliki nilai yang kecil ketika dikurangi dengan today
# 1 artinya customernya skor recency nya kecil, karena sudah lama tidak aktif ( data customer recency dikurangi dengan today )
# recency berbanding terbalik karena semakin dekat dengan today artinya customer aktif paling terbaru
```

## Segmentation menggunakan pandas cluster (pendekatan percentile) (pd.cut)


#### Dari gambar diatas, kita hanya mengelompokkan jadi 4 jenis saja dengan ketentuan label maksimal ialah 4 :
1. Champion Customers : RecencyScore : 3-4, FrequencyScore : 3-4, MonetaryScore : 3-4
2. Loyal Customer : RecencyScore : 2-4, FrequencyScore : 2-4, MonetaryScore : 4
3. Need Attention : RecencyScore : 1-4 , FrequencyScore : 1-3 ,  MonetaryScore : 3-4
4. At Risk : RecencyScore : 1-3, FrequencyScore : 1-3, MonetaryScore : 1-2
5. Contoh loyal customer : 234, 224, 434

#### Referensi : https://blog.rsquaredacademy.com/customer-segmentation-using-rfm-analysis/

## Pembuatan laporan stop sampai disini, code kebawah ga perlu dimasukkan, code diatas dimasukkan tapi sampai bagian k-means dengan code frequency vs monetary : data_rfm["Frequency"], y = data_rfm['MonetaryValue']

```python
# Anggap Anda memiliki DataFrame customer_data dengan kolom RecencyScore, FrequencyScore, dan MonetaryScore.

def customer_segment(row):
    # Mendapatkan nilai dari setiap kolom
    recency = row['RecencyScore']
    frequency = row['FrequencyScore']
    monetary = row['MonetaryScore']

    if recency in [3, 4] and frequency in [3, 4] and monetary in [3, 4]:
        return "Champion Customers"
    elif recency in [2, 3, 4] and frequency in [2, 3, 4] and monetary == 4:
        return "Loyal Customer"
    elif recency in [1, 2, 3, 4] and frequency in [1, 2, 3] and monetary in [3, 4]:
        return "Need Attention"
    elif recency in [1, 2, 3] and frequency in [1, 2, 3] and monetary in [1, 2]:
        return "At Risk"
    else:
        return "At Risk"
```

```python
# Menerapkan fungsi customer_segment ke setiap baris DataFrame
customer_data['CustomerSegment'] = customer_data.apply(lambda row: customer_segment(row), axis=1)
```

```python
customer_data.CustomerSegment.value_counts()
```

Dalam kode di atas, untuk menerapkan sebuah fungsi bernama customer_segment ke setiap baris dalam DataFrame customer_data menggunakan metode apply dengan axis=1. Fungsi tersebut kemudian menghasilkan segmen pelanggan berdasarkan skor RFM yang ada pada setiap baris

Dengan menerapkan fungsi customer_segment ke setiap baris DataFrame, Anda dapat mengidentifikasi segmen-segmen pelanggan berdasarkan skor RFM mereka. Hal ini membantu Anda memahami distribusi pelanggan di setiap segmen dan merencanakan strategi pemasaran yang lebih efektif dan disesuaikan dengan masing-masing segmen.
